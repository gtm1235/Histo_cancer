##Time to define our CNN model

```python
model = models.resnet18(pretrained=True)
model
```

```python
for param in model.parameters():
  param.requires_grad = False
  
from collections import OrderedDict
fc = nn.Sequential(OrderedDict([
    ('hidden_1', nn.Linear(512, 512)),
    ('relu1', nn.ReLU()),
    ('dropout1', nn.Dropout(0.5)),
    ('hidden_2', nn.Linear(512, 128)),
    ('relu2', nn.ReLU()),
    ('dropout2', nn.Dropout(0.5)),
    ('hidden_4', nn.Linear(128, 2)),
]))

model.fc = fc
for param in model.fc.parameters():
    param.requires_grad = True

print(model)
```

##Add Optimizer and Loss finctions

```python
criterion = nn.BCELoss()
optimizer = optim.SGD(model.fc.parameters(), lr=.01)
sigmoid = nn.Sigmoid()
ps_output= nn.LogSoftmax(dim=1)
```

#Training Loop

```python
steps = 0
running_loss = 0
epochs = 150
print_every = 50
valid_loss_min = np.Inf

for e in range(epochs):
  model.train()
  model.to(device)
  counter = 0
  for images, labels in train_loader:
    steps += 1
    print(steps)
    images, labels = images.to(device), labels.to(device)
    
    optimizer.zero_grad()
    
    output = model.forward(images)
    loss = criterion(sigmoid(output), labels)
    loss.backward()
    optimizer.step()
    
    running_loss += loss.item()
    #print(counter)
    #counter += 1
    if steps % print_every == 0:
      model.eval()
      
      
      accuracy = 0
      test_loss = 0
      with torch.no_grad():
        for images, labels in valid_loader:
          images, labels = images.to(device), labels.to(device)
          print(steps)
        
          output = model(images)
          test_loss += criterion(output, labels).item()
        
          ps = torch.exp(ps_output(output))
          equality = (labels.data == ps.max(1)[1])
          accuracy += equality.type_as(torch.FloatTensor()).mean()
        
      print("Epoch: {}/{}.. ".format(e+1, epochs),
            "Training Loss: {:.3f}.. ".format(running_loss/print_every),
            "Validation Loss: {:.3f}.. ".format(test_loss/len(valid_loader)),
            "Validation Accuracy: {:.3f}".format(accuracy/len(valid_loader)))
      
      # save model if validation loss has decreased
      valid_loss = test_loss/len(valid_loader)
      
      if valid_loss <= valid_loss_min:
        
        checkpoint = {'batch_size': 64,
                      'valid_transforms': valid_transforms,
                      'train_transforms': train_transforms,
                      'model': model,
                      'classifier': fc,
                      'criterion': criterion,
                      'optimizer': optimizer.state_dict(),
                      'state_dict': model.state_dict()
                      }
        
        print('Validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.
              format(valid_loss_min, valid_loss))
        
        torch.save(checkpoint,'/content/drive/My Drive/histo_data/models/model_checkpoint.pth')
        valid_loss_min = valid_loss
      
      running_loss = 0
      model.train()
        
        
```
